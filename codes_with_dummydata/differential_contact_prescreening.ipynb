```python
import pandas as pd
import numpy as np
```

```python
%%time
# Load holo system data #
# TxN matrix where T=Number of datapoints and N= number of features 
holodata=pd.read_csv("newdatahead.csv")
```

```python
%%time
# Load data #
apodata=pd.read_csv("newdatatail.csv")
```

```python
%%time
# Load data #
mergeddata=pd.read_csv("newdatamerged.csv")
```

```python
# Select those features whose distances fall within a contact threshold in atleast 75% frames of the trajectory #
subset1 = apodata.loc[:, (apodata < 5).sum() >= 0.75 * apodata.shape[0]]
apo_list = subset1.columns.tolist()
print(apo_list)
subset1
```

```python
# Select those features whose distances fall within a contact threshold in atleast 75% frames of the trajectory #
subset2 = holodata.loc[:, (holodata < 5).sum() >= 0.75 * holodata.shape[0]]
holo_list = subset2.columns.tolist()
print(holo_list)
subset2
```

```python
# Create a list of all identified edges #
tmp = apo_list + holo_list
totallist = []
[totallist.append(x) for x in tmp if x not in totallist] # Remove duplicates 
print(len(apo_list))
print(len(holo_list))
print(len(totallist))
```

```python
## FIND UNIQUE EDGES OF APO AND HOLO SYSTEMS ##
uniqueapo = []
[uniqueapo.append(x) for x in apo_list if x not in holo_list]
print(len(uniqueapo))
print(uniqueapo)

uniqueholo = []
[uniqueholo.append(x) for x in holo_list if x not in apo_list]
print(len(uniqueholo))
print(uniqueholo)

uniquelist = uniqueapo + uniqueholo
len(uniquelist)
```

```python
# Remove features = distances between two adjacent residues #
new_list = []
count=0
diff=[]
for x in uniquelist:
    first, second = x.split('-')
    if abs(int(first) - int(second)) == 1:
        count+=1
        uniquelist.remove(x)
        new_list.append(x)
print(count)
print(new_list)
print(len(uniquelist))
```

```python
# Select features and labels #
features = mergeddata[uniquelist]
target = mergeddata['labels']
# Combine features and target into a single DataFrame
df_to_save = pd.concat([features, target], axis=1)
# Save to CSV file
df_to_save.to_csv("merged.csv", index=False)
df_to_save.head()
```
